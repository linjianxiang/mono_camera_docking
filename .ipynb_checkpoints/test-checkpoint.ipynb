```python
import cv2
import numpy as np
import glob
from matplotlib import pyplot as plt
 
```

```python
top_dir = '/home/linjian/dataset/tum_image/sequence_30/images/00050.jpg'
img = cv2.imread(top_dir)
plt.imshow(img)

print(img.shape)
resized = cv2.resize(img,(1280//2,1024//2), interpolation = cv2.INTER_AREA)
print(resized.shape)
plt.imshow(resized)
```

```python
ref_dir = '/home/linjian/dataset/tum_image/sequence_30/images/'
dest_dir = '/home/linjian/dataset/tum_image/sequence_30/resized_images/'
filelist = glob.glob(ref_dir+'/*.jpg')
filelist = sorted(filelist)
```

```python
counter = 0
length = len(filelist)


for i in range(0,length,4):
    img = cv2.imread(filelist[i])
    resized = cv2.resize(img,(1280//2,1024//2), interpolation = cv2.INTER_AREA)
    
    cv2.imwrite(dest_dir+''+str(counter)+'.jpg', resized) 
    counter=counter +1
    
print(counter)

# cv2.imwrite(filename, img) 
```
